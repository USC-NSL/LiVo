```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

%matplotlib inline
```

```python
input_dir = "/home/lei/rajrup/KinectStream/Scripts/ablation/parallel_decoding_overhead/"

data_no_decoding_df = pd.read_csv(
    os.path.join(input_dir, "gpu_log_no_decoding_py_log.csv")
)

data_with_decoding_df = pd.read_csv(
    os.path.join(input_dir, "gpu_log_with_decoding_py_log.csv")
)
```

```python
# Remove rows where decoder_util(%) is 0
data_no_decoding_df = data_no_decoding_df[data_no_decoding_df["decoder_util(%)"] > 2]
data_with_decoding_df = data_with_decoding_df[
    data_with_decoding_df["decoder_util(%)"] > 2
]
```

```python
# Statistics for no decoding

print("Statistics with no decoding")
print(f"Mean - gpu_util(%): {data_no_decoding_df['gpu_util_percent'].mean():.2f}")
print(f"Std - gpu_util(%): {data_no_decoding_df['gpu_util_percent'].std():.2f}")
print(f"Median - gpu_util(%): {data_no_decoding_df['gpu_util_percent'].median():.2f}")

print()

print(f"Mean - mem_util(%): {data_no_decoding_df['mem_util_percent'].mean():.2f}")
print(f"Std - mem_util(%): {data_no_decoding_df['mem_util_percent'].std():.2f}")
print(f"Median - mem_util(%): {data_no_decoding_df['mem_util_percent'].median():.2f}")

print()

print(f"Mean - memory_used(MB): {data_no_decoding_df['mem_used_mib'].mean():.2f}")
print(f"Std - memory_used(MB): {data_no_decoding_df['mem_used_mib'].std():.2f}")
print(f"Median - memory_used(MB): {data_no_decoding_df['mem_used_mib'].median():.2f}")

print()

print(f"Mean - decoder_util(%): {data_no_decoding_df['decoder_util(%)'].mean():.2f}")
print(f"Std - decoder_util(%): {data_no_decoding_df['decoder_util(%)'].std():.2f}")
print(
    f"Median - decoder_util(%): {data_no_decoding_df['decoder_util(%)'].median():.2f}"
)
```

```python
# Statistics for with decoding

print("Statistics with decoding")
print(f"Mean - gpu_util(%): {data_with_decoding_df['gpu_util_percent'].mean():.2f}")
print(f"Std - gpu_util(%): {data_with_decoding_df['gpu_util_percent'].std():.2f}")
print(f"Median - gpu_util(%): {data_with_decoding_df['gpu_util_percent'].median():.2f}")

print()

print(f"Mean - mem_util(%): {data_with_decoding_df['mem_util_percent'].mean():.2f}")
print(f"Std - mem_util(%): {data_with_decoding_df['mem_util_percent'].std():.2f}")
print(f"Median - mem_util(%): {data_with_decoding_df['mem_util_percent'].median():.2f}")

print()

print(f"Mean - memory_used(MB): {data_with_decoding_df['mem_used_mib'].mean():.2f}")
print(f"Std - memory_used(MB): {data_with_decoding_df['mem_used_mib'].std():.2f}")
print(f"Median - memory_used(MB): {data_with_decoding_df['mem_used_mib'].median():.2f}")

print()

print(f"Mean - decoder_util(%): {data_with_decoding_df['decoder_util(%)'].mean():.2f}")
print(f"Std - decoder_util(%): {data_with_decoding_df['decoder_util(%)'].std():.2f}")
print(
    f"Median - decoder_util(%): {data_with_decoding_df['decoder_util(%)'].median():.2f}"
)
```

```python
# Difference in gpu_util(%)

print("Difference in gpu_util(%)")
print(
    f"Mean: {data_with_decoding_df['gpu_util_percent'].mean() - data_no_decoding_df['gpu_util_percent'].mean():.2f}"
)

print("Difference in mem_util(%)")
print(
    f"Mean: {data_with_decoding_df['mem_util_percent'].mean() - data_no_decoding_df['mem_util_percent'].mean():.2f}"
)

print("Difference in mem_used(MB)")
print(
    f"Mean: {data_with_decoding_df['mem_used_mib'].mean() - data_no_decoding_df['mem_used_mib'].mean():.2f}"
)

print("Difference in decoder_util(%)")
print(
    f"Mean: {data_with_decoding_df['decoder_util(%)'].mean() - data_no_decoding_df['decoder_util(%)'].mean():.2f}"
)
```
