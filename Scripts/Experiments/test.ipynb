```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Sample data
data = {
    'boxes': [
        {'med': 1, 'q1': 0.5, 'q3': 1.5, 'whislo': 0.2, 'whishi': 2.7, 'fliers': []},
        {'med': 3, 'q1': 2, 'q3': 4, 'whislo': 1, 'whishi': 6, 'fliers': []}
    ],
    'whiskers': [
        (0.2, 2), (0.5, 2.7),
        (1, 3), (1.5, 4),
        (2, 5), (2.7, 6)
    ],
    'caps': [
        (0.2, 0.5),
        (2, 2.7),
        (1, 1.5),
        (4, 4),
        (5, 5),
        (6, 6)
    ]
}

# Creating a figure and axis
fig, ax = plt.subplots()

# Creating boxplot using bxp
bp = ax.bxp(data['boxes'], patch_artist=True)

# Setting face color for boxes
for box in bp['boxes']:
    box.set(facecolor='lightblue')  # Change 'lightblue' to your desired color

# Setting color for median lines
for median in bp['medians']:
    median.set(color='red')  # Change 'red' to your desired color

# Create legend rectangle patches
box_patch = mpatches.Patch(color='lightblue', label='Box')
median_patch = mpatches.Patch(color='red', label='Median')

# Add legend with rectangle patches
plt.legend(handles=[box_patch, median_patch], loc='upper center', ncol=2, bbox_to_anchor=(0.5, 1.02), fontsize=14)

plt.show()
```
