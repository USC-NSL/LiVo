```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
import os

%matplotlib inline
```

```python
ground = "with_ground"
# ground = "no_ground"
seq_name = f"160906_band2_{ground}"
logID = 1
metric = "pointssim"

data_dir = f"/datassd/pipeline_cpp/client_tiled/ablation/bitrate_study/{seq_name}/"

data_dir = os.path.join(data_dir, f"comp_o3d_{metric}", f"log{logID}")
```

```python
color_qp = 0
data_dict = {}

for depth_qp in range(0, 36, 5):
    pssim_path = os.path.join(data_dir, f"3D_cqp{color_qp}_d_yuvqp{depth_qp}_pssim.csv")
    if not os.path.exists(pssim_path):
        raise FileNotFoundError(f"{pssim_path} not found")
    data = pd.read_csv(pssim_path)
    data_dict[depth_qp] = data

    avg_pssim_geo_ba = np.mean(data["PSSIM_GEO_BA"])
    avg_pssim_color_ba = np.mean(data["PSSIM_COLOR_BA"])
    print(
        f"d_qp={depth_qp}, avg_pssim_geo_dist_gt={avg_pssim_geo_ba:.2f}, avg_pssim_color_dist_gt={avg_pssim_color_ba:.2f}"
    )

metric_names = ["PSSIM_GEO_BA", "PSSIM_COLOR_BA"]
exp_names = ["PSSIM_GEO_DIST_GT", "PSSIM_COLOR_DIST_GT"]

for metric_name in metric_names:
    fig, ax = plt.subplots(1, 1, figsize=(10, 5))
    for depth_qp in range(0, 36, 5):
        data = data_dict[depth_qp]
        ax.plot(data["Frame"], data[metric_name], label=f"d_qp={depth_qp}")

    ax.set_title(exp_names[metric_names.index(metric_name)])
    ax.set_xlabel("Frame ID", fontsize=15)
    ax.set_ylabel("PSSIM", fontsize=15)
    ax.legend()
    ax.set_ylim(0, 100)
    plt.show()
```

```python
ground = "with_ground"
# ground = "no_ground"
seq_name = f"160906_band2_{ground}"
logID = 1
metric = "pointssim"

data_dir = f"/datassd/pipeline_cpp/client_tiled/ablation/bitrate_study/{seq_name}/"

data_dir = os.path.join(data_dir, f"bitrate", f"log{logID}")
```

```python
color_qp = 0
depth_bitrate_dict = {}

for depth_qp in range(0, 36, 5):
    bitrate_path = os.path.join(
        data_dir,
        f"dbitrate_cqp{color_qp}_d_yuvqp{depth_qp}_fps30_s_nocull_c_nocull.csv",
    )
    if not os.path.exists(bitrate_path):
        raise FileNotFoundError(f"{bitrate_path} not found")
    data = pd.read_csv(bitrate_path)
    depth_bitrate_dict[depth_qp] = data
```

```python
# Remove frame id less than 500
for depth_qp in range(0, 36, 5):
    data = depth_bitrate_dict[depth_qp]
    data = data[data["FrameID"] > 500]
    data = data[data["Bitrate(in kbps)"] > 100.0]
    depth_bitrate_dict[depth_qp] = data
```

```python
# Plot depth bitrate
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
for depth_qp in range(0, 36, 5):
    data = depth_bitrate_dict[depth_qp]
    ax.plot(data["FrameID"], data["Bitrate(in kbps)"], label=f"d_qp={depth_qp}")
ax.set_title("Depth Bitrate")
ax.set_xlabel("Frame ID", fontsize=15)
ax.set_ylabel("Bitrate (kbps)", fontsize=15)
ax.set_ylim(0, 1.6e5)
ax.legend()
plt.show()
```

```python
depth_qp = 0
color_bitrate_dict = {}

for color_qp in range(0, 51, 5):
    bitrate_path = os.path.join(
        data_dir,
        f"cbitrate_cqp{color_qp}_d_yuvqp{depth_qp}_fps30_s_nocull_c_nocull.csv",
    )
    if not os.path.exists(bitrate_path):
        raise FileNotFoundError(f"{bitrate_path} not found")
    data = pd.read_csv(bitrate_path)
    color_bitrate_dict[color_qp] = data
```

```python
# Fix color qp at 0 and vary depth qp
# Metric vs Depth QP

################# PSSIM Geometric ####################

metric = "pointssim"
ground = "with_ground"
seq_names = [
    f"160906_band2_{ground}",
    f"170307_dance5_{ground}",
    f"170915_office1_{ground}",
    f"160906_pizza1_{ground}",
    f"170915_toddler4_{ground}",
]
logIDs = [1, 0, 2, 1, 0]


color_qp = 0
depth_qps = list(range(0, 36, 5))
metric_names = ["PSSIM_GEO_BA"]
metric_names_exp = ["PSSIM_GEO_DIST_GT"]
fig, ax = plt.subplots(1, 1, figsize=(10, 5))

markers = [
    "o",
    "s",
    "v",
    "^",
    "D",
    "X",
    "P",
    "p",
    "H",
    "h",
    "*",
    "d",
    "x",
    "+",
    "|",
    "_",
]

markers = markers[: len(seq_names)]
for seq_name, logID, marker in zip(seq_names, logIDs, markers):
    data_path = os.path.join(
        "/datassd/pipeline_cpp/client_tiled/ablation/bitrate_study",
        seq_name,
        f"comp_o3d_{metric}",
        f"log{logID}",
    )

    data_dict = {}
    for depth_qp in depth_qps:
        data_filepath = os.path.join(
            data_path,
            f"3D_cqp{color_qp}_d_yuvqp{depth_qp}_pssim.csv",
        )
        if not os.path.exists(data_filepath):
            raise FileNotFoundError(f"{data_filepath} not found")

        data = pd.read_csv(data_filepath)
        column_names = list(data.columns)[1:]
        data_dict[depth_qp] = data

    for metric_name, metric_name_exp in zip(metric_names, metric_names_exp):
        metric_dict = {}
        for depth_qp in depth_qps:
            data = data_dict[depth_qp]
            data = data[metric_name]
            metric_dict[depth_qp] = data.mean(axis=0)
        print(f"{seq_name}, Avg. PSSIM Geo: ", metric_dict)
        ax.plot(depth_qps, list(metric_dict.values()), label=seq_name, marker=marker)

ax.set_ylim(0.0, 100.0)
ax.set_ylabel("PSSIM GEO Metric")
ax.set_xlabel("Depth QP")
ax.set_title("Fix color qp = 0, vary depth qp")
ax.legend(loc="lower right")

################# PSSIM Color ####################
metric_names = ["PSSIM_COLOR_BA"]
metric_names_exp = ["PSSIM_COLOR_DIST_GT"]
fig, ax = plt.subplots(1, 1, figsize=(10, 5))

for seq_name, logID, marker in zip(seq_names, logIDs, markers):
    data_path = os.path.join(
        "/datassd/pipeline_cpp/client_tiled/ablation/bitrate_study",
        seq_name,
        f"comp_o3d_{metric}",
        f"log{logID}",
    )

    data_dict = {}
    for depth_qp in depth_qps:
        data_filepath = os.path.join(
            data_path,
            f"3D_cqp{color_qp}_d_yuvqp{depth_qp}_pssim.csv",
        )
        if not os.path.exists(data_filepath):
            raise FileNotFoundError(f"{data_filepath} not found")

        data = pd.read_csv(data_filepath)
        column_names = list(data.columns)[1:]
        data_dict[depth_qp] = data

    for metric_name, metric_name_exp in zip(metric_names, metric_names_exp):
        metric_dict = {}
        for depth_qp in depth_qps:
            data = data_dict[depth_qp]
            data = data[metric_name]
            metric_dict[depth_qp] = data.mean(axis=0)
        print(f"{seq_name}, Avg. PSSIM Color: ", metric_dict)
        ax.plot(depth_qps, list(metric_dict.values()), label=seq_name, marker=marker)

ax.set_ylim(0.0, 100.0)
ax.set_ylabel("PSSIM COLOR Metric")
ax.set_xlabel("Depth QP")
ax.set_title("Fix color qp = 0, vary depth qp")
ax.legend(loc="lower right")

################# Depth Bitrate ####################

fig, ax = plt.subplots(1, 1, figsize=(10, 5))

for seq_name, logID, marker in zip(seq_names, logIDs, markers):
    data_path = os.path.join(
        "/datassd/pipeline_cpp/client_tiled/ablation/bitrate_study",
        seq_name,
        "bitrate",
        f"log{logID}",
    )

    data_dict = {}
    for depth_qp in depth_qps:
        data_filepath = os.path.join(
            data_path,
            f"dbitrate_cqp{color_qp}_d_yuvqp{depth_qp}_fps30_s_nocull_c_nocull.csv",
        )
        if not os.path.exists(data_filepath):
            raise FileNotFoundError(f"{data_filepath} not found")

        data_dict[depth_qp] = pd.read_csv(data_filepath)

    metric_dict = {}
    for depth_qp in depth_qps:
        data = data_dict[depth_qp]
        data = data[data["FrameID"] > 500]
        data = data[data["Bitrate(in kbps)"] > 100.0]
        data_dict[depth_qp] = data
        metric_dict[depth_qp] = (
            data["Bitrate(in kbps)"].mean(axis=0) / 1.0e3
        )  # Convert to Mbps
    print(f"{seq_name}, Avg. Bitrate: ", metric_dict)
    ax.plot(depth_qps, list(metric_dict.values()), label=seq_name, marker=marker)

ax.set_ylim(0.0, 200.0)
ax.set_ylabel("Depth Bitrate (Mbps)")
ax.set_xlabel("Depth QP")
ax.set_title("Fix color qp = 0, vary depth qp")
ax.legend(loc="upper right")
```
